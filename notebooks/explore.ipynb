## Cleaning

A continuacion limpieza del dataset, eliminando los nulos de Age

```python
df.Age[ (df.Sex=='female' )& (df.Age.isnull()==1)]=df[df.Sex=='female'].Age.median()
```

```python
df.Age[ (df.Sex=='male' )& (df.Age.isnull()==1)]=df[df.Sex=='male'].Age.median()
```

```python
df2.Age[ (df2.Sex=='female' )& (df2.Age.isnull()==1)]=df2[df2.Sex=='female'].Age.median()
```

```python
df2.Age[ (df2.Sex=='male' )& (df2.Age.isnull()==1)]=df2[df2.Sex=='male'].Age.median()
```
